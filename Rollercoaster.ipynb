```python
df = df.loc[~df.duplicated(subset = ['Coaster_Name', 'Location', 'Opening_Date'])] \
.reset_index(drop = True).copy()
```

```python
df['Year_Introduced'].value_counts()
```

```python
ax = df['Year_Introduced'].value_counts() \
    .head(10) \
    .plot(kind = 'bar', title = 'Top 10 years of construction of new attractions')
ax.set_xlabel('Year')
ax.set_ylabel('Quantity ')
```

```python
ax2 = df['Speed_Mph'].plot(kind='hist', 
                           bins=30, 
                           title='Slide Speed in mph')
ax2.set_xlabel('Speed in (mph)')
```

```python
ax2 = df['Speed_Mph'].plot(kind='kde', 
                           title='Slide speed in mph')
ax2.set_xlabel('Speed in (Mph)')
```

```python
df.plot(kind='scatter',
       x='Speed_Mph',
       y='Height_Ft',
       title='Dependence of speed on height')
plt.show()
```

```python
sns.scatterplot(x='Speed_Mph',
                y='Height_Ft',
                hue='Year_Introduced',
                data=df)
```

```python
sns.pairplot(df, vars=['Year_Introduced','Speed_Mph',
                       'Height_Ft', 'Inversions', 'Gforce'],
             hue='Type_Main')
plt.show()
```

```python
df_corr = df[['Year_Introduced','Speed_Mph',
   'Height_Ft', 'Inversions', 'Gforce']].dropna().corr()
```

```python
sns.heatmap(df_corr, annot=True)
```

```python
ax = df.query('Location != "Other"') \
      .groupby('Location')['Speed_Mph'] \
      .agg(['mean','count']) \
      .query('count > 10') \
      .sort_values('mean')['mean'] \
      .plot(kind='barh',
            figsize=(12,5),
            title='Average speed of the slides at each location')
ax.set_xlabel('average speed')
plt.show()
```
